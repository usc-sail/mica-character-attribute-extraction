```python
import requests
import pprint
import bs4
```

```python
session = requests.session()
```

```python
url = "https://en.wikipedia.org/w/api.php"

params = {
    "action": "opensearch",
    "namespace": "0",
    "search": "The Matrix (1999)",
    "limit": "5",
    "format": "json"
}

response = session.get(url=url, params=params)
data = response.json()

print(data)
```

```python
url = "https://en.wikipedia.org/w/api.php"

params = {
    "action": "query",
    "prop": "revisions",
    "titles": "The_Matrix",
    "rvslots": "*",
    "rvprop": "content",
    "formatversion": "2",
    "format": "json"
}

response = session.get(url=url, params=params)
data = response.json()
```

```python
url = "https://en.wikipedia.org/w/api.php"

params = {
    "action": "parse",
    "page": "THE MATRIX (1999)",
    "format": "json",
    "redirects": True,
    "section": 1,
    "prop": "text"
}

response = session.get(url=url, params=params)
data = response.json()

soup = bs4.BeautifulSoup(data["parse"]["text"]["*"], "html.parser")
```

```python
print(soup.prettify())
```

```python
print(soup.text.strip())
```
